```python
from sklearn import preprocessing
lable_encoder = preprocessing.LabelEncoder()
dataset["State_Name"]= lable_encoder.fit_transform(dataset["State_Name"])
dataset["District_Name"]= lable_encoder.fit_transform(dataset["District_Name"])
dataset["Season"]= lable_encoder.fit_transform(dataset["Season"])
dataset["Crop"]= lable_encoder.fit_transform(dataset["Crop"])
dataset
```

```python
dataset.isnull().sum()
```

```python
a=dataset.dropna()
a
```

```python
a=a.head(1000)
a
```

```python
a.isnull().sum()
```

```python
x=a.iloc[:,:-1].values
y=a.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=588)
x_train
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
y_pred=regressor.predict(x_test)
y_pred
```

```python
print("Train Accuracy:", regressor.score(x_train,y_train))
print("Test Accuracy:", regressor.score(x_test,y_pred))
```
