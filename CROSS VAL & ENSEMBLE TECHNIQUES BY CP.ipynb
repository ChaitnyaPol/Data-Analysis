# Using label encoder for convert object to interger

```python
from sklearn.preprocessing import LabelEncoder
A1=LabelEncoder()
data["Education"]=A1.fit_transform(data["Education"])
#data["City"]=A1.fit_transform(data["City"])
data["Gender"]=A1.fit_transform(data["Gender"])
data["EverBenched"]=A1.fit_transform(data["EverBenched"])
```

# Data slicing

```python
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
y
```

# Data imbalancing using SMOTE

```python
from imblearn.over_sampling import SMOTE
S= SMOTE()
x_data,y_data=S.fit_resample(x,y)
```

```python
from collections import Counter
```

```python
Counter(y_data)
```

# Split the data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_data,y_data,test_size=0.2,random_state=11)
```

# Cross Validation LogisticRegression

```python
from sklearn.model_selection import KFold
kf=KFold(n_splits=5,random_state=11,shuffle=True)
kf.get_n_splits(x_data)
print(kf)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
from sklearn.linear_model import LogisticRegression
L=LogisticRegression()
```

```python
scores=cross_val_score(L,x_train,y_train,cv=kf)
y_pred=cross_val_predict(L,x_test,y_test,cv=kf)
print(Counter(y_train))
print('@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@')
print(Counter(y_test))
print('%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%')
print(scores)
print('////////////////////////////////////////////////')
print(y_pred)
print('><><><><><><><><><><><><><><><><><><><><><><><><')
print(np.mean(scores)*100)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)*100
```

# Ensemble Techniques

```python
y_pred1=L.predict(x_test)
y_pred1
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred1,y_test)*100
```

# StratifiedKFold

```python
from sklearn.model_selection import StratifiedKFold
sf=StratifiedKFold(n_splits=5,random_state=10,shuffle=True)
sf.get_n_splits(x_data,y_data)
for train_data,test_data in sf.split(x_data,y_data):
    x_train,x_test=x_data[train_data],x_data[test_data]
    y_train,y_test=y_data[train_data],y_data[test_data]
    scores=cross_val_score(L,x_train,y_train,cv=kf)
    y_pred1=cross_val_predict(L,x_test,y_test,cv=kf)
    print(Counter(y_test))
    print(scores)
    print('//////////////////////////')
    print(y_pred1)
    print('><><><><><><><><><><><><><><><>')
    print(np.mean(scores)*100)
    print('#############################')
    print(x_test,y_test)
    print('><><><><><><><><><><><><><><><><><><<>><><><><><><><><><><><><><><><><<><><><><><><><><><>>><><><><><><>>><><>')
```
