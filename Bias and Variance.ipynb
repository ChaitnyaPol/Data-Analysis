```python
data=pd.read_csv(r"C:\Users\comp\Downloads\heart.csv")
data
```

```python
from sklearn.preprocessing import LabelEncoder
L1=LabelEncoder()
data["Sex"]=L1.fit_transform(data["Sex"])
data["ChestPainType"]=L1.fit_transform(data["ChestPainType"])
data["RestingECG"]=L1.fit_transform(data["RestingECG"])
data["ExerciseAngina"]=L1.fit_transform(data["ExerciseAngina"])
data["ST_Slope"]=L1.fit_transform(data["ST_Slope"])
data
```

```python
x1=data.iloc[:,:-1].values
y1=data.iloc[:,-1].values
```

```python
from imblearn.over_sampling import SMOTE
s11=SMOTE()
x_data1,y_data1=s11.fit_resample(x1,y1)
```

```python
from sklearn.model_selection import train_test_split
x_train1,x_test1,y_train1,y_test1=train_test_split(x_data1,y_data1,test_size=0.2,random_state=11)
```

```python
from sklearn.linear_model import LogisticRegression
l1=LogisticRegression()
l1.fit(x_train1,y_train1)
```

```python
y_train_bal=l1.predict(x_train1)

from sklearn.metrics import accuracy_score
ac2=accuracy_score(y_train_bal,y_train1)*100
ac2
```

```python
y_test_bal=l1.predict(x_test1)

from sklearn.metrics import accuracy_score
ac3=accuracy_score(y_test_bal,y_test1)*100
ac3
```

```python
var_train=y_data1.var()

sse=np.mean(np.mean(y_train_bal-y_train1)**2)
bias1=sse-var_train
bias1
```

```python
var_test=y_data1.var()

sse=np.mean(np.mean(y_test_bal-y_test)**2)
bias11=sse-var_test
bias11
```
